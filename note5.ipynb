#  Key Equations and Intuition

## 1. Objective Function (Social Welfare)

DICE views climate change in framework of economic growth theory, sandtard <u>Ramsey model</u>, society invests in captial goods, thereby reducing consumpyion today, to increase consumption in the future. DICE includes **climate investments**, which are analogous to capital investments in the standard model.

The DICE model optimizes **social welfare $W$**, defined as the discounted sum of the population-weighted utility of per capita consumption:

$$
W = \sum_{t=1}^{T_{\max}} V\big(c(t), L(t), R(t)\big) = \sum_{t=1}^{T_{\max}} U\big(c(t)\big) \cdot L(t) \cdot R(t)
$$

---

### Variables:

| Variable        | Description                                             |
|-----------------|---------------------------------------------------------|
| $W$             | Social welfare (total discounted utility over time)    |
| $t$             | Time index (year or period)                             |
| $T_{\max}$      | Final time period in the model                          |
| $V$             | Instantaneous social welfare at time $t$               |
| $U(c(t))$       | Utility function of per capita consumption at time $t$ |
| $c(t)$          | Per capita consumption at time $t$                      |
| $L(t)$          | Population at time $t$                                  |
| $R(t)$          | Discount factor at time $t$, $R(t) = (1+\rho)^{-t}$    |
| $\rho$          | Pure rate of social time preference (discount rate)   |
| $\alpha$        | Generational inequality aversion parameter (curvature of utility) |


discount rate determines how much climate mitigation the model considers optimal

---

### Utility function $U(c)$:

The model uses a **constant relative risk aversion (CRRA)** utility function:

$$
U(c) = \frac{c^{1-\alpha}}{1 - \alpha}
$$

where:

- $\alpha$ controls the degree of inequality aversion between generations.

---

### Intuition:

- Social welfare $W$ sums over time the utility of consumption weighted by population and discounted to reflect time preference.
- The discount factor $R(t)$ makes future utilities less valuable than present ones.
- $\alpha$ tunes how society values consumption differences over time (inequality aversion).
- If $\alpha \to 1$, the utility function becomes logarithmic:

$$
U(c) = \ln(c)
$$


where  

$$
U(C_t,L_t) = L_t \frac{(C_t/L_t)^{1-\alpha} - 1}{1-\alpha}
$$

- **Intuition:** Trade-off between present vs. future consumption (discounted utility).

# Net Output in the DICE

Net output, $Q(t)$, is defined as **gross output reduced by damages from climate change and mitigation (abatement) costs**:

$$
Q(t) = \Omega(t) \big[1 - \Lambda(t)\big] Y(t)
$$

where:

- $Q(t)$ = Net output, i.e., output available after accounting for damages and mitigation.
- $\Omega(t)$ = Scaling factor or productivity factor
- $\Lambda(t)$ = abatement cost, fractional losses due to climate damages and mitigation costs
- $Y(t)$ = Gross output, is allocated between total consumption and gross investmen

### Additional notes:

- The **gross output $Y(t)$** is modeled as a **Cobb-Douglas function** of capital, labor, and technology
- Labor input is proportional to population.
- Capital accumulates according to an optimized savings rate.
---
# Damage Function in the DICE

The damage function $\Omega(t)$ is defined as:

$$
\Omega(t) = \frac{1}{1 + D(t)}
$$

where the damage term $D(t)$ is:

$$
D(t) = \phi_1 \, T_{AT}(t) + \phi_2 \, \big[T_{AT}(t)\big]^2
$$


### Explanation:

- $\Omega(t)$ represents the fraction of gross output remaining after damages
- $D(t)$ represents the damage fraction caused by temperature increase.
- $T_{AT}(t)$ is the atmospheric temperature increase
- $\phi_1$ and $\phi_2$ are parameters that determine the linear and quadratic effects of temperature on damages, respectively.

---

The damage function captures how rising temperatures reduce the productive capacity of the economy in the DICE model.

## 4. Climate Module

**Carbon Stock:**

$$
M_{t+1} = \Phi M_t + E_t
$$

**Temperature:**

$$
T_{t+1} = T_t + \xi (F_t - \lambda T_t)
$$

- **Intuition:** Captures inertia of carbon cycle and delayed warming.

---

## 5. Damage Function

$$
Y_t^{net} = \frac{Y_t}{1 + \psi_1 T_t + \psi_2 T_t^2}
$$

- **Intuition:** Output falls as temperature rises, with nonlinear damages.

---

## 6. Social Cost of Carbon

$$
SCC_t = - \frac{\partial W / \partial E_t}{\partial W / \partial C_t}
$$

- **Intuition:** Consumption-equivalent loss of 1 extra ton of CO₂ today.

---

## 7. Optimal Carbon Price

Optimal carbon price \(P_t\) equals SCC:

$$
P_t = SCC_t
$$

- **Intuition:** Internalizes externality → efficient climate policy.

---

#  Economic Intuition

- **Early Action:** Cheaper than waiting, because damages compound.
- **Smooth Path:** Optimal carbon price rises steadily, avoids shocks.
- **Discount Rate Critical:** Lower discount rate ⇒ higher SCC ⇒ stronger policy.
- **Marginal Damages:** Updated damage function ⇒ higher SCC ⇒ more aggressive mitigation.
